# 서울시 CCTV 현황 데이터 분석

### 1. 서울통계서비스 사이트에서 '주민등록인구(구별)' 통계에서 2018년도 데이터를 다운 받아서 '서울시 인구현황.txt' 파일로 저장한다.

### 2. 파일에서 데이터를 읽어 컬럼이 '자치구', '인구수', '내국인', '외국인', '고령자' 인 데이터 프레임을 만든다.(pop)

```python
import numpy as np
import pandas as pd
```

```python
pop = pd.read_csv('서울시인구현황.txt', sep='\t',thousands=',',usecols=[1,3,6,9,13],
                  skiprows=[0,1,2,3],names=['자치구', '인구수', '내국인', '외국인', '고령자'])
pop
```

### 3. 외국인 비율과 고령자 비율을 추가한다.


```python
pop['외국인 비율'] = pop['외국인']/pop['인구수']*100
```

```python
pop['고령자 비율'] = pop['고령자']/pop['인구수']*100
```

```python
pop.head()
```

### 4. 서울시 자치구 년도별 CCTV 설치 현황 데이터를 다운 받아서 '서울시 자치구 년도별 CCTV 설치 현황.xlsx' 파일로 저장한후, 데이터를 읽서 데이터 프레임을 만든다.(cctv)

```python
cctv = pd.read_excel("서울시 자치구 년도별 CCTV 설치 현황.xlsx")
cctv
```

### 5. '기관명' 컬럼 데이터의 빈칸을 없애고, 컬럼명을 '자치구'로 변경한다.

```python
cctv['기관명'].unique()
```

```python
cctv['기관명'] = ['강남구', '강동구', '강북구', '강서구', '관악구', '광진구', '구로구',
       '금천구', '노원구', '도봉구', '동대문구', '동작구', '마포구', '서대문구',
       '서초구', '성동구', '성북구', '송 파 구', '양천구', '영등포구', '용산구',
       '은평구', '종로구', '중구', '중랑구']
cctv
```

```python
cctv['기관명'] = [x.replace(' ','') for x in cctv['기관명'].unique()]
```

```python
cctv.rename(columns = {'기관명' : '자치구'}, inplace = True)  #inplace = True 는 원본데이터를 고치라는 것
cctv
```

### 6. 2015년 이전의 데이터를 합하여 '2015년 이전' 컬럼을 만들고, 최근 3개년간의 데이터를 합하여 '최근 3개년' 컬럼을 만든 후, '최근 증가율' 컬럼을 만든다.

```python
### nan은 0으로 바꾸고 인티저로
cctv[['2011년 이전','2012년','2013년']] = cctv[['2011년 이전','2012년','2013년']].fillna(0).astype(int)
cctv
```

```python
cctv['2015년 이전'] = cctv[['2011년 이전','2012년','2013년','2014년','2015년']].sum(axis=1)
#cctv['2015년 이전'] = cctv,iloc[:,2:7].sum(axis=1)
cctv.head()
```

```python
cctv['최근 3개년'] = cctv[['2016년','2017년','2018년']].sum(axis=1)
cctv.head()
```

```python
cctv['최근 증가율'] = cctv['최근 3개년']/cctv['2015년 이전']*100
cctv.head()
```

### 7. '자치구', '소계', '최근 증가율' 컬럼을 제외하고 나머지는 모두 삭제한다.

```python
cctv.drop(['2011년 이전','2012년','2013년','2014년','2015년'],axis=1,inplace=True)
cctv.head()
```

```python
del cctv['2016년']
cctv.head()
```

```python
del cctv['2017년']
del cctv['2018년']
del cctv['2015년 이전']
del cctv['최근 3개년']
cctv.head()
```

### 8. 서울시 인구 데이터와 cctv 데이터를 합치고, '자치구'를 index로 만든다.

```python
pop.head()
```

```python
df = pd.merge(cctv,pop,on='자치구')
df.head()
```

```python
df.set_index('자치구',inplace=True)
df.head()
```

### 9. CCTV 현황을 다양한 형태로 시각화하여 보여준다.

```python
# Matplotlib 사용과 한글 처리를 위한 Code
import matplotlib as mpl 
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False 
mpl.rc('font', family='Malgun Gothic')   
%matplotlib inline

plt.rcParams["figure.figsize"] = (12,6)
plt.rcParams["font.size"] = 14
```

### -자치구별 cctv 갯수 (내림차순으로)

```python
df['소계'].sort_values()
```

```python
df1 = df.sort_values(by=['소계'],axis=0,ascending=False)  #ascending=False 내림차순
df1
```

```python
df['소계'].sort_values().plot(kind='barh',grid=True,figsize=(12,12))
plt.show()
```

### 자치구별 인당 CCTV 갯수 (내림차순으로)

```python
df['인당 CCTV'] = df['소계']/df['인구수']
df['인당 CCTV'].sort_values().plot(kind='barh',grid=True, figsize=(12,12))
plt.show()
```

```python
df['최근 증가율'].sort_values().plot(kind='barh',grid=True, figsize=(12,12))
plt.show()
```

### -인구수와 CCTv 댓수의 산점도

```python
plt.scatter(df['인구수'], df['소계'], s=50)
plt.xlabel('인구수')
plt.ylabel('CCTV 대수')
plt.show()
```

### -인구수와 CCTV 댓수의 산점도 및 회귀선

```python
fp1 = np.polyfit(df['인구수'],df['소계'],1)
fp1           ##기울기와 y절편이 나옴
```

```python
f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000,101)
```

```python
plt.figure(figsize=(12,10))
plt.scatter(df['인구수'], df['소계'], s=50)
plt.plot(fx,f1(fx), ls = 'dashed', lw = 3, color='red')
plt.xlabel('인구수')
plt.ylabel('CCTV 대수')
plt.grid()
plt.show()
```

### -인구수와 CCTV 댓수의 산점도 및 회귀선 및 자치구명

```python
df.head()
```

```python
df['오차']=np.abs(df['소계'] - f1(df['인구수']))
df.head()
```

```python
df_sort = df.sort_values(by='오차',ascending=False)
df_sort.head(10)
```

```python
plt.figure(figsize=(14,10))
plt.scatter(df['인구수'],df['소계'],c = df['오차'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')

for n in range(10):
    plt.text(df_sort['인구수'][n]*1.02, df_sort['소계'][n]*0.98, df_sort.index[n], fontsize=15)
    
plt.xlabel('인구수')
plt.ylabel('CCTV대수')
plt.colorbar()
plt.grid()
plt.show()
```

### -고령자 비율 vs 인당 CCTV

```python
plt.figure(figsize=(14,10))
plt.scatter(df['고령자 비율'],df['인당 CCTV'], s=50)
    
plt.xlabel('고령자 비율')
plt.ylabel('인당 CCTV')
plt.grid()
plt.show()
```

### -외국인 비율 vs 인당 CCTV

```python
plt.figure(figsize=(14,10))
plt.scatter(df['외국인 비율'],df['인당 CCTV'], s=50)
    
plt.xlabel('외국인 비율')
plt.ylabel('인당 CCTV')
plt.grid()
plt.show()
```

```python
fp3 = np.polyfit(df['외국인 비율'],df['인당 CCTV'],1)
f3 = np.poly1d(fp3)
fx = np.linspace(0.5, 0.5,101)
```

```python
plt.figure(figsize=(12,10))
plt.scatter(df['외국인 비율'], df['인당 CCTV'], s=50)
plt.plot(fx, f3(fx), ls='dashed', lw=3, color='m')
plt.xlabel('외국인 비율')
plt.ylabel('인당 CCTV')
plt.grid()
plt.show()
```
