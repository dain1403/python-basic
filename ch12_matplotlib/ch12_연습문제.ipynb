```python
# Matplotlib 사용과 한글 처리를 위한 Code
import matplotlib as mpl 
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False 
mpl.rc('font', family='Malgun Gothic')   
%matplotlib inline

plt.rcParams["figure.figsize"] = (12,6)
plt.rcParams["font.size"] = 14
```

```python
#1번
import csv
f = open('cafe_2year.csv', 'r', encoding='utf-8')
lines = csv.reader(f)

header = next(lines)

date_data = []
numCafe_data = []
monthCount = 1

for line in lines :
    if monthCount % 3 == 0 :
        date_data.append(line[0])
        numCafe_data.append(line[1])
    monthCount += 1

plt.bar(date_data, numCafe_data)

plt.title('전국의 월별 까페수',color='green')
plt.xlabel('년월')
plt.ylabel('카페수')

plt.show()
```

```python
# 2번
f = open('doctor_2019.csv', 'r', encoding='utf-8')
lines = csv.reader(f)

header = next(lines)

area_data = ['서울','부산','대구','인천','대전','광주','울산']
general_practitioner_num_data=[0,0,0,0,0,0,0] # 서울,부산..순,,일반의 수
intern_num_data=[0,0,0,0,0,0,0]
resident_num_data=[0,0,0,0,0,0,0]
specialist_num_data=[0,0,0,0,0,0,0]

for line in lines :
    if line[0] == '서울':
        general_practitioner_num_data[0] += int(line[2])
        intern_num_data[0] += int(line[3])
        resident_num_data[0] += int(line[4])
        specialist_num_data[0] += int(line[5])
    elif line[0] == '부산':
        general_practitioner_num_data[1] += int(line[2])
        intern_num_data[1] += int(line[3])
        resident_num_data[1] += int(line[4])
        specialist_num_data[1] += int(line[5])
    elif line[0] == '대구':
        general_practitioner_num_data[2] += int(line[2])
        intern_num_data[2] += int(line[3])
        resident_num_data[2] += int(line[4])
        specialist_num_data[2] += int(line[5])
    elif line[0] == '인천':
        general_practitioner_num_data[3] += int(line[2])
        intern_num_data[3] += int(line[3])
        resident_num_data[3] += int(line[4])
        specialist_num_data[3] += int(line[5])
    elif line[0] == '대전':
        general_practitioner_num_data[4] += int(line[2])
        intern_num_data[4] += int(line[3])
        resident_num_data[4] += int(line[4])
        specialist_num_data[4] += int(line[5])
    elif line[0] == '광주':
        general_practitioner_num_data[5] += int(line[2])
        intern_num_data[5] += int(line[3])
        resident_num_data[5] += int(line[4])
        specialist_num_data[5] += int(line[5])
    elif line[0] == '울산':
        general_practitioner_num_data[6] += int(line[2])
        intern_num_data[6] += int(line[3])
        resident_num_data[6] += int(line[4])
        specialist_num_data[6] += int(line[5])
        
print(header[0],header[2],header[3],header[4],header[5])
for i in range(7):
    print(area_data[i],'\t',general_practitioner_num_data[i],'\t',
        intern_num_data[i],'\t', resident_num_data[i],'\t',
        specialist_num_data[i])
f.close()
```

```python
# 3번
import csv
f1 = open('doctor2.csv', 'w', encoding='utf-8',newline='')
fwr = csv.writer(f1)
fwr.writerow(['지역','일반의 수','레지던트 수','전문의 수'])
for i in range(7):
    fwr.writerow([area_data[i],general_practitioner_num_data[i],intern_num_data[i], resident_num_data[i],specialist_num_data[i]])
    
print('doctor2.csv 파일 쓰기 완료!')
f1.close()
```

```python
#4번
f2 = open('doctor2.csv', 'r', encoding='utf-8')
lines = csv.reader(f2)

header = next(lines)

area_data1 = []
doctor_num_data = []

for line in lines :
    #print(line)
    area_data1.append(line[0])
    doctor_num_data.append(int(line[4]))
    
plt.scatter(area_data1,doctor_num_data)
plt.title('서울 및 광역시 종합병원 전문의 수',color='green')
plt.xlabel('시')
plt.ylabel('의사 수')

plt.show()
f2.close()
```

```python
# 5번
f3 = open('doctor2.csv', 'r', encoding='utf-8')
lines = csv.reader(f3)

header = next(lines)

area_data2 = []
general_data = []
intern_data=[]
resident_data=[]
specialist_data=[]

for line in lines :
    area_data2.append(line[0])
    general_data.append(int(line[1]))
    intern_data.append(int(line[2]))
    resident_data.append(int(line[3]))
    specialist_data.append(int(line[4]))
    
plt.plot(area_data2, general_data, label='일반의', color='red', linestyle='-', marker='o')
plt.plot(area_data2, intern_data, label='인턴', color='blue', linestyle='--', marker='*')
plt.plot(area_data2, resident_data, label='레지던트', color='yellow', linestyle=':', marker='s')
plt.plot(area_data2, specialist_data, label='전문의', color='purple', linestyle='-.', marker='d')

plt.title('서울 및 광역시 종합병원 의사 수',color = "green")
plt.xlabel('시')
plt.ylabel('의사 수')

plt.show()
f3.close()
```

```python
# 6번
division_data = ['일반의','인턴','레지던트','전문의']
doctor_percent = []

f4 = open('doctor2.csv', 'r', encoding='utf-8')
lines = csv.reader(f4)

header = next(lines)
total = 0
for i in range(4):
    total += int(line[i+1])
    
for i in range(4):
    percent = int(line[i+1])/total
    doctor_percent.append(percent)
    
plt.pie(doctor_percent, explode=(0, 0.1, 0, 0), labels=division_data, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('서울 종합병원 의사 유형별 분포',color = "green")

plt.show()
f4.close()
```
