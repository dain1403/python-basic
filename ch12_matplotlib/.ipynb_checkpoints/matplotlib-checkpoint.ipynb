```python
# Matplotlib 사용과 한글 처리를 위한 Code
import matplotlib as mpl 
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False 
mpl.rc('font', family='Malgun Gothic')   
%matplotlib inline
```

```python
#12-4
xdata = ['안지영', '홍지수', '황예린']
ydata1 = [90, 85, 88]
ydata2 = [83, 88, 91]
plt.plot(xdata, ydata1, label='국어')
plt.plot(xdata, ydata2, label='영어')
plt.legend(loc='upper center')   #범례

plt.title('세명 학생의 국어, 영어 성적')

plt.show()
```

```python
#12-5 Line Style
xdata = ['안지영', '홍지수', '황예린']
ydata1 = [90, 85, 88]
ydata2 = [83, 88, 91]
ydata3 = [85, 97, 78]
ydata4 = [92, 88, 82]

plt.plot(xdata, ydata1, label='국어', color='red', linestyle='-', marker='o') # -는 실선
plt.plot(xdata, ydata2, label='영어', color='#00ffff', linestyle='--', marker='x') # -- 점선
plt.plot(xdata, ydata3, label='수학', color='magenta', linestyle='-.', marker='s')  
plt.plot(xdata, ydata4, label='사회', color='#444444', linestyle=':', marker='d')

plt.title('세명 학생의 네 과목 성적')
plt.legend(loc='best')

plt.show()
```

```python
#12-5 Line Style
xdata = ['안지영', '홍지수', '황예린']
ydata1 = [90, 85, 88]
ydata2 = [83, 88, 91]
ydata3 = [85, 97, 78]
ydata4 = [92, 88, 82]

plt.plot(xdata, ydata1, 'ro-', label='국어')    # 간단하게 변경
plt.plot(xdata, ydata2, 'cx--', label='영어') 
plt.plot(xdata, ydata3, 'ms-.', label='수학')  
plt.plot(xdata, ydata4, 'bd:', label='사회')

plt.title('세명 학생의 네 과목 성적')
plt.legend(loc='best')

plt.show()
```

```python
#12-5 Line Style
xdata = ['안지영', '홍지수', '황예린']
ydata1 = [90, 85, 88]
ydata2 = [83, 88, 91]
ydata3 = [85, 97, 78]
ydata4 = [92, 88, 82]

plt.plot(xdata, ydata1, 'ro-', label='국어', linewidth = 3)  # 두껍게  
plt.plot(xdata, ydata2, 'cx--', label='영어', linewidth = 3) 
plt.plot(xdata, ydata3, 'ms-.', label='수학', linewidth = 3)  
plt.plot(xdata, ydata4, 'bd:', label='사회', linewidth = 3)

plt.title('세명 학생의 네 과목 성적')
plt.legend(loc='best')

plt.show()
```

```python
#12-5 Line Style
xdata = ['안지영', '홍지수', '황예린']
ydata1 = [90, 85, 88]
ydata2 = [83, 88, 91]
ydata3 = [85, 97, 78]
ydata4 = [92, 88, 82]

plt.plot(xdata, ydata1, 'ro-', label='국어', linewidth = 3, markersize = 10) 
plt.plot(xdata, ydata2, 'cx--', label='영어', linewidth = 3, markersize = 10) 
plt.plot(xdata, ydata3, 'ms-.', label='수학', linewidth = 3, markersize = 10)  
plt.plot(xdata, ydata4, 'bd:', label='사회', linewidth = 3, markersize = 10)

plt.title('세명 학생의 네 과목 성적')
plt.legend(loc='best')

plt.show()
```

```python
# 12-6
xdata = list(range(30))
ydata = []

for x in xdata:
    y = 2*x
    ydata.append(y)
    
plt.title('y=2x')
plt.plot(xdata, ydata,color='magenta', linewidth = 3)
plt.xlim(-1,30)
plt.ylim(-2,60)
plt.xticks(list(range(0,30,2)))
plt.yticks(list(range(0,60,5)))

plt.grid(True)
plt.show()
```

```python
# sin, cos 그리기
import numpy as np

x = np.arange(0, 2*numpy.pi, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1,linewidth = 3)
plt.plot(x, y2,linewidth = 3)
plt.show()
```

```python
xs = np.arange(0, 2*numpy.pi, 0.1)
ysin = np.sin(xs)
ycos = np.cos(xs)

plt.plot(xs, ysin, label = 'Sine curve')
plt.plot(xs, ycos, label = 'Cosine curve')
plt.title('삼각함수 그래프')
plt.legend()
plt.xticks(np.arange(0, 2*numpy.pi*0.01,np.pi/2))

plt.axhline(y=0, color='k',linewidth = 0.5)
plt.axhline(y=1, color='k',linewidth = 0.5)
plt.axhline(y=-1, color='k',linewidth = 0.5)
plt.axvline(x=0, color='k',linewidth = 0.5)
plt.axvline(x=np.pi, color='k',linewidth = 0.5)
plt.axvline(x=2*np.pi, color='k',linewidth = 0.5)

plt.show()
```

```python
# sigmoid함수 y = 1/(1+e^(-x)) 

xs = np.linspace(-3, 3, 601)
ys = 1 / (1 + np.exp(-xs))

plt.plot(xs, ys)
plt.title('시그모이드 그래프')

plt.axhline(y=0, color='k',linewidth = 0.5)
plt.axhline(y=1, color='k',linewidth = 0.5)
plt.axvline(x=0, color='k',linewidth = 0.5)

plt.grid()
plt.show()
```
